```python
import numpy as np
import pandas as pd
```

```python
data=pd.read_csv('BostonHousing.csv')
data
```

```python
from seaborn import boxplot
boxplot(x=data['dis'])
```

```python
q1=data.quantile(.25)
q3=data.quantile(.75)
iqr=q3-q1
```

```python
orig_data=data.copy()
```

```python
data
```

Before removing outliers we had 506 rows

```python
#this method is applied for all the columns
data=data[~((data<(q1 - 1.5*iqr)) | (data>(q3 + 1.5*iqr))).any(axis=1)]
```

after removing outliers we have 268 rows
